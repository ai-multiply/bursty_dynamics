## Calculate BP and MC. 

We can calculate the BP and the MC per subject id using the calculate_scores function. You will need to input the data, subject id, and date column. 

```python
# calculate scores
score_df = calculate_scores(df, subject_id = 'eid', time_col = 'event_dt')
score_df.head()
```

We can view a scatter plot of BP against MC by using the argument 'scatter'.

```python
score_df = calculate_scores(df, subject_id = 'eid', time_col = 'event_dt', scatter=True)
```

We can view histogram of the BP and MC using the argument 'hist'. <br>
hist (str or None): <br>
- True: Plot histograms for both BP and MC.<br>
- "BP": Plot histogram for BP only.<br>
- "MC": Plot histogram for MC only.<br>
- "Both": Plot histograms for both BP and MC on the same plot.<br>
- None: Do not plot any histograms.<br>


```python
score_df = calculate_scores(df, subject_id='eid', time_col='event_dt', scatter=True, hist=True)
```

```python
score_df = calculate_scores(df, subject_id='eid', time_col='event_dt', scatter=True, hist='Both')
```

## Detect trains

We can detect trains using the train_detection function. you will need to input the data, subject_id, time_col, and max_iet. <br>
Max_iet is the maximum distance between consecutive events in a train (in days).

Note: train_id = 0 is not a train. They represent events that are not in a train. 

```python
train_df = train_detection(df, subject_id = 'eid', time_col = 'event_dt', max_iet = 30)
```

```python
train_df.head()
```

## Train information

We can use the calculate_train_info fucntion to find more information about the trains.

```python
train_info = calculate_train_info(train_df, subject_id = 'eid', time_col = 'event_dt')
```

```python
train_info.head()
```

## Calculate scores of trains

We can calculate the scores of the trains detected using the calculate_scores_train function. <br>
min_event_n: the mimimum number of events that should be in a train. 
If a number x is given to the argument min_event_n, the resulting data will only show trains that have at least x number of events in the train, excluding train_id = 0. 

```python
train_score = calculate_scores_train(train_df, subject_id = 'eid', time_col ='event_dt', min_event_n= 3)
train_score.head()
```

We can display scatter plot and histogram using scatter and hist arguments. <br>

hist (str or None): <br>
- True: Plot histograms for both BP and MC.<br>
- "BP": Plot histogram for BP only.<br>
- "MC": Plot histogram for MC only.<br>
- "Both": Plot histograms for both BP and MC on the same plot.<br>
- None: Do not plot any histograms.<br>

```python
train_score = calculate_scores_train(train_df, subject_id='eid', time_col='event_dt', min_event_n=5, scatter=True, hist=True)
train_score.head()
```

```python
train_score = calculate_scores_train(train_df, subject_id='eid', time_col='event_dt', min_event_n=5, scatter=True, hist='Both')
train_score.head()
```
